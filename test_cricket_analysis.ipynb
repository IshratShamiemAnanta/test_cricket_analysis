```python
#importing the necessary python libraries
import pandas as pd
import numpy as np
```

```python
#reading the wickets csv file
df = pd.read_csv("wickets.csv", encoding = 'unicode_escape')
```

```python
#displaying the first 10 rows of the dataframe
display(df.head(10))
```

<b>Player</b> : Name of the players who has bowled in test cricket<br> 
<b>Span</b> : Duration of time the player was active in playing cricket<br> 
<b>Mat</b> : Number of Matches the player has played<br> 
<b>Inns</b> : Number of innings the player has bowled in<br> 
<b>Balls</b> : Number of Balls the player has bowled<br> 
<b>Runs</b> : Number of runs the player has conceded<br> 
<b>BBI</b> : Best Bowling figure in an Innings<br> 
<b>BBM</b> : Best Bowling figure in a Match<br> 
<b>Ave</b> : Average which is calculated by dividing the numbers of runs conceded by the number of wickets taken<br> 
<b>Econ</b> : Economy Rate, which is the average number of runs conceded for each over bowled<br> 
<b>SR</b> : Strike Rate, which is calculated as total balls bowled / wickets taken by bowler<br> 
<b>5</b> : Number of times the player has taken 5 wickets in a match<br> 
<b>10</b> : Number of times the player has taken 10 wickets in a match<br>

```python
#renaming the column names appropriately
df = df.rename(columns={"Mat":"Match", "Inns":"Innings","Wkts":"Wickets","BBI":"Best Bowling figure/Innings","BBM":"Best Bowling figure/Match","Ave":"Average","Econ":"Economy","SR":"Strike Rate", "5":"5 wickets/match","10":"10 wickets/match"})

display(df.head())
```

```python
#removing columns named 'Best Bowling figure/Innings' and 'Best Bowling figure/Match' from the dataframe
df = df.drop('Best Bowling figure/Innings', axis = 1)
df = df.drop('Best Bowling figure/Match', axis = 1)

display(df.head())
```

```python
#splitting the country part from the player's name in the 'Player' column
df_player = df['Player'].str.split("(", expand = True) 

display(df_player.head())
```

```python
df = pd.concat([df, df_player], axis = 1)

display(df.head())
```

```python
df = df.drop('Player', axis = 1)
display(df.head())
```

```python
df = df.rename(columns={0:'Player',1:'Country'})
display(df.head())
```

```python
df['Country'] = df['Country'].str.replace(")","")
display(df.head())
```

```python
new_col_sequence = ['Player', 'Country','Span','Match','Innings','Balls','Runs','Wickets','Average','Economy','Strike Rate','5 wickets/match','10 wickets/match']

df = df[new_col_sequence]
display(df.head())
```

```python
#function to check if 'ICC' is in the 'Country' column
def icc_check(x):
    if "ICC" in x:
        return 1
    else:
        return 0
```

```python
df['Played for ICC'] = df['Country'].apply(icc_check)
display(df.head())
```

```python
df['Played for ICC'].value_counts()
```

**Observation** <br>
It can be deduced that 5 players played for ICC.


```python
df['Country'] = df['Country'].str.replace("ICC/","")
df['Country'] = df['Country'].str.replace("/ICC","")
display(df.head())
```

```python
df['Country'].value_counts()
```

**Observations** <br>
1. There are 10 different countries present in this dataset. <br>
2. There are 18 Australian Bowlers present in this dataset. <br>
3. There is only 1 Bangladeshi player present in this dataset.

```python
#slicing the start and end years from the 'Span' column
df['start_year'] = df['Span'].str[0:4]
df['end_year'] = df['Span'].str[5:]
display(df.head())
```

```python
#converting the 'start_year' and 'end_year' column into integer data type
df['start_year'] = df['start_year'].astype('int')
df['end_year'] = df['end_year'].astype('int')
```

```python
#calculating the years played by a player
df['Years Played'] = df['end_year'] - df['start_year']
display(df.head())
```

```python
#removing the 'start_year' and 'end_year' columns from the dataframe
df = df.drop('start_year', axis = 1)
df = df.drop('end_year', axis = 1)

display(df.head())
```

```python
#sorting players with respect to years they played
df = df.sort_values(by="Years Played", ascending = False)
display(df.head())
display(df.tail())
```

**Observations** <br>
1. Imran Khan of Pakistan had played for the longest period of time i.e. 21 years <br>
2. GP Swann of England had played for the shortest period of time i.e. 5 years

```python
#sorting players with respect to their Economy rate
df = df.sort_values(by="Economy")
display(df.head())
```

**Observation** <br>

```python
#sorting players with respect to Strike Rate
df = df.sort_values(by="Strike Rate")
display(df.head())
```

```python
#sorting players with respect to their Average
df = df.sort_values(by="Average")
display(df.head())
```
