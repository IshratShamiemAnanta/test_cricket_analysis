```python
#importing the necessary python libraries
import pandas as pd
import numpy as np
```

```python
#reading the wickets csv file
df = pd.read_csv("wickets.csv", encoding = 'unicode_escape')
```

```python
#displaying the first 10 rows of the dataframe
display(df.head(10))
```

<b>Player</b> : Name of the players who has bowled in test cricket<br> 
<b>Span</b> : Duration of time the player was active in playing cricket<br> 
<b>Mat</b> : Number of Matches the player has played<br> 
<b>Inns</b> : Number of innings the player has bowled in<br> 
<b>Balls</b> : Number of Balls the player has bowled<br> 
<b>Runs</b> : Number of runs the player has conceded<br> 
<b>BBI</b> : Best Bowling figure in an Innings<br> 
<b>BBM</b> : Best Bowling figure in a Match<br> 
<b>Ave</b> : Average which is calculated by dividing the numbers of runs conceded by the number of wickets taken<br> 
<b>Econ</b> : Economy Rate, which is the average number of runs conceded for each over bowled<br> 
<b>SR</b> : Strike Rate, which is calculated as total balls bowled / wickets taken by bowler<br> 
<b>5</b> : Number of times the player has taken 5 wickets in a match<br> 
<b>10</b> : Number of times the player has taken 10 wickets in a match<br>
